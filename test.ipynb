```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import PIL
import ddddocr
import numpy as np
```

```python
ocr = ddddocr.DdddOcr(show_ad=False)
```

```python
n=3
img = cv2.imread('captcha.png', 0)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.erode(img, np.ones((n, n), np.uint8))
img = cv2.bitwise_not(img)

cv2.imwrite('captcha_test.png', img)
with open('captcha_test.png', 'rb') as f:
    image_bytes = f.read()
plt.imshow(cv2.imread('captcha_test.png', 0), cmap='gray')

ocr.classification(image_bytes)
```
